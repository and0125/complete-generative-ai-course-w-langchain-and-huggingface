```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = (
    OllamaEmbeddings(model="llama3.2") # uses Llama model by default
)
```

```python
result1 = embeddings.embed_documents(
    ["This is a test document.", 
     "This is another test document."] # note: this should be a list of strings
    )
```
